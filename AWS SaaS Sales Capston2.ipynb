```python
# Function to normalize city names by removing special characters
def normalize_city_name(city_name):
    # Normalize the city name to decompose special characters
    normalized = unicodedata.normalize('NFKD', city_name)
    # Encode to ASCII bytes, ignore errors, and decode back to string
    return normalized.encode('ASCII', 'ignore').decode('ASCII')

# Apply the normalization function to the 'City' column
data['City'] = data['City'].apply(normalize_city_name)

# Dictionary with the city name corrections
corrections = {
    'Novogorod': 'Nizhny Novgorod',
    'Krakow': 'Krakow',
    'Lisboa': 'Lisbon'
}

# Update city names with corrections in the normalized column
data['City'] = data['City'].replace(corrections)

# Display the updated DataFrame
data['City'].head()
```

4.6.2. Data Outliers

```python
# Box plots to identify outliers
def plot_boxplots(data):
    numerical_cols = data.select_dtypes(include=['float64', 'int64']).columns
    plt.figure(figsize=(20, 15))
    for i, col in enumerate(numerical_cols, 1):
        plt.subplot(4, 3, i)
        sns.boxplot(x=data[col])
        plt.title(f'Box plot of {col}')
    plt.tight_layout()
    plt.show()

# Plotting box plots for outliers
plot_boxplots(data)
```

```python
# Function to calculate the number of outliers in each column and return as DataFrame
def count_outliers_df(data):
    outlier_counts = {'Column': [], 'Outlier Count': []}
    
    for col in data.select_dtypes(include=['float64', 'int64']).columns:
        Q1 = data[col].quantile(0.25)
        Q3 = data[col].quantile(0.75)
        IQR = Q3 - Q1
        
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        
        # Count outliers
        outliers = data[(data[col] < lower_bound) | (data[col] > upper_bound)]
        outlier_counts['Column'].append(col)
        outlier_counts['Outlier Count'].append(len(outliers))
        
    return pd.DataFrame(outlier_counts)

# Calculate the number of outliers for each numerical column and convert to DataFrame
outlier_counts_df = count_outliers_df(data)

# Show the DataFrame
outlier_counts_df
```

6.3.2. Statistical Tests For Countries of ContactMatcher Toward Median Discount

```python
# Filter data for 'ContactMatcher'
contact_matcher_data = data[data['Product'] == 'ContactMatcher']

# Calculate the median discount for each country
median_discount_by_country = contact_matcher_data.groupby('Country')['Discount'].median().reset_index()

# Get the top 15 countries with the highest median discount
top_15_countries = median_discount_by_country.nlargest(15, 'Discount')['Country'].tolist()

# Get the top 15 countries with the highest median discount
lowest_5_countries = median_discount_by_country.nsmallest(5, 'Discount')['Country'].tolist()

# Combine the selected countries
selected_countries = top_15_countries + lowest_5_countries

# Mann-Whitney Test (for comparing two groups)
# Example: Comparing Australia and Japan
australia_discounts = contact_matcher_data[contact_matcher_data['Country'] == 'Australia']['Discount']
japan_discounts = contact_matcher_data[contact_matcher_data['Country'] == 'Japan']['Discount']

mann_whitney_stat, mann_whitney_p = mannwhitneyu(australia_discounts, japan_discounts)
print(f"Mann-Whitney Test between Australia and Japan: Stat={mann_whitney_stat}, p-value={mann_whitney_p}")

# Kruskal-Wallis Test (for comparing more than two groups)
discounts_by_country = [contact_matcher_data[contact_matcher_data['Country'] == country]['Discount'] for country in selected_countries]

kruskal_stat, kruskal_p = kruskal(*discounts_by_country)
print(f"Kruskal-Wallis Test for selected countries: Stat={kruskal_stat}, p-value={kruskal_p}")
```

6.3.3. Create Definitions And Features of Top 15 Countries For Graph Analysis

```python
# Revised code to find the top 15 countries by median discount and highest sales

# Function to find the group with the highest median discount
def find_highest_discount_group(df, group_by_column, top_n=None):
    grouped_data = df.groupby(group_by_column)['Discount'].median().reset_index()
    if top_n:
        grouped_data = grouped_data.nlargest(top_n, 'Discount')
    highest_discount_group = grouped_data.loc[grouped_data['Discount'].idxmax()]
    return highest_discount_group, grouped_data

# Analyze 'Country' and get top 15
highest_discount_country, top_countries = find_highest_discount_group(contact_matcher_data, 'Country', top_n=15)

# Calculate the total sales for the top 15 countries
top_countries_sales = contact_matcher_data[contact_matcher_data['Country'].isin(top_countries['Country'])] \
    .groupby('Country')['Sales'].sum().reset_index()

# Calculate the total profit for the top 15 countries
top_countries_profit = contact_matcher_data[contact_matcher_data['Country'].isin(top_countries['Country'])] \
    .groupby('Country')['Profit'].sum().reset_index()

# Calculate the median profit percentage for the top 15 countries
top_countries_percentage = contact_matcher_data[contact_matcher_data['Country'].isin(top_countries['Country'])] \
    .groupby('Country')['Profit Percentage'].median().reset_index()

# Calculate the overall profit percentage for the top 15 countries
top_countries_overall = contact_matcher_data[contact_matcher_data['Country'].isin(top_countries['Country'])] \
    .groupby('Country').agg({'Sales': 'sum', 'Profit': 'sum'}).reset_index()

# Calculate the overall profit percentage
top_countries_overall['Overall Profit Percentage'] = (top_countries_overall['Profit'] / top_countries_overall['Sales']) * 100

# Ensure countries are in the same order
ordered_top_countries = top_countries['Country'].tolist()
top_countries_profit = top_countries_profit.set_index('Country').loc[ordered_top_countries].reset_index()
top_countries_sales = top_countries_sales.set_index('Country').loc[ordered_top_countries].reset_index()
top_countries_percentage = top_countries_percentage.set_index('Country').loc[ordered_top_countries].reset_index()
top_countries_overall = top_countries_overall.set_index('Country').loc[ordered_top_countries].reset_index()

# Define a consistent color palette
palette = sns.color_palette("mako", len(ordered_top_countries))
```

6.4.1. Find Out The Distribution Normality of ContactMatcher Product Across All Industries

```python
# Assuming contact_matcher_data is your DataFrame
industries = contact_matcher_data['Industry'].unique()

# Perform normality tests
normality_results = []

for industry in industries:
    industry_data = contact_matcher_data[contact_matcher_data['Industry'] == industry]['Discount']
    
    # Shapiro-Wilk test
    shapiro_stat, shapiro_p = shapiro(industry_data)
    
    # Kolmogorov-Smirnov test
    ks_stat, ks_p = kstest(industry_data, 'norm', args=(industry_data.mean(), industry_data.std()))
    
    normality_results.append({
        'Industry': industry,
        'Shapiro-Wilk Test Statistic': shapiro_stat,
        'Shapiro-Wilk p-value': shapiro_p,
        'Kolmogorov-Smirnov Test Statistic': ks_stat,
        'Kolmogorov-Smirnov p-value': ks_p
    })

normality_df = pd.DataFrame(normality_results)
normality_df
```

```python
# Interpretation
for index, row in normality_df.iterrows():
    if row['Shapiro-Wilk p-value'] < 0.05 or row['Kolmogorov-Smirnov p-value'] < 0.05:
        print(f"Industry {row['Industry']} does not follow a normal distribution.")
    else:
        print(f"Industry {row['Industry']} follows a normal distribution.")
```

- Insights

1. The results from the Shapiro-Wilk and Kolmogorov-Smirnov tests indicate that the discount data for all industries does not follow a normal distribution. This conclusion is based on the p-values from both tests being less than 0.05 for all industries, leading to the rejection of the null hypothesis of normality.
2. Implication for Analysis, consider using non-parametric statistical methods for further analysis:

| **Statistic Approach** | **Interpretation**                                        |
|:---------------:|:-------------------------------------------------------------:|

| Comparative Analysis      | Use non-parametric tests such as the Mann-Whitney U test or Kruskal-Wallis test for comparing groups instead of t-tests or ANOVA.     |

6.5.11. Cohort Analysis of Returning Customers
- The cohort analysis aimed to filter key customers contributing to significant cohort drops in retention.
- The cohort analysis supports the line chart’s insight that there are fluctuations in customer retention over time, with some periods showing notable drops.
- The heat map provides a more granular view, showing exactly when these drops occur and which cohorts are most affected.

```python
# Ensure 'Order Date' is in datetime format
contact_matcher_data['Order Date'] = pd.to_datetime(contact_matcher_data['Order Date'])

# Extract the year and month of the first purchase to define cohorts
contact_matcher_data['YearMonth'] = contact_matcher_data['Order Date'].dt.to_period('M')

# Assign cohorts based on the first purchase
contact_matcher_data['Cohort_Month'] = contact_matcher_data.groupby('Customer')['YearMonth'].transform('min')

# Convert 'Cohort_Month' to period format (monthly)
contact_matcher_data['Cohort_Month'] = contact_matcher_data['Cohort_Month'].astype('period[M]')

# Convert to quarterly periods for analysis
contact_matcher_data['Current_Order_Q'] = contact_matcher_data['Order Date'].dt.to_period('Q')
contact_matcher_data['Cohort_Q'] = contact_matcher_data['Cohort_Month'].apply(lambda x: x.asfreq('Q'))

# Calculate the difference between the transaction month and the cohort start month
contact_matcher_data['Period_Number'] = (contact_matcher_data['Current_Order_Q'] - contact_matcher_data['Cohort_Q']).apply(lambda x: x.n)

# Calculate customer retention per cohort
contact_matcher_cohort = contact_matcher_data.groupby(by=['Cohort_Q', 'Period_Number']).agg({'Customer': 'nunique'}).reset_index()

# Pivot the table to get cohort periods as columns
pivot_cohort = contact_matcher_cohort.pivot_table(index='Cohort_Q', columns='Period_Number', values='Customer')

# Convert the values to percentages
percentage_pivot_cohort = (pivot_cohort.div(pivot_cohort[0], axis=0) * 100).round(2)

# Fill NaN values with zeros for visualization
percentage_pivot_cohort = percentage_pivot_cohort.fillna(0)

# Display the cohort table
print(percentage_pivot_cohort)

# Set the size of the subplot/axes
fig, (ax_top, ax_bottom) = plt.subplots(nrows=2, ncols=1, figsize=(12, 10), gridspec_kw={'height_ratios': (20, 1), 'hspace': 0.3})

# Heatmap
ax_top = sns.heatmap(
    data=percentage_pivot_cohort,
    cmap='YlGnBu',                             # color palette
    ax=ax_top,                                 # subplot for heatmap
    cbar_ax=ax_bottom,                         # subplot for color bar
    cbar_kws={'orientation': 'horizontal'},    # change orientation of color bar
    annot=True,                                # display annotations
    fmt='.2f',                                 # format annotations to 2 decimal places
    linewidths=2                               # border between cells
)

# Show plot
plt.show()
```

6.6.2. CLV (Customer Lifetime Value) Approach In The Analysis
- Using the CLV approach helps in understanding and maximizing the profitability of customers over their entire relationship with the company. Here are the key reasons why this approach is used.

```python
# Filter data for the selected key countries and industries
selected_countries = ['Australia', 'Japan', 'Canada', 'United Kingdom', 'Brazil', 'Spain']
selected_industries = ['Transportation', 'Finance', 'Manufacturing', 'Energy', 'Tech', 'Healthcare']

# Segment data by customer within these countries & industries
filtered_data = contact_matcher_data[
    (contact_matcher_data['Country'].isin(selected_countries)) &
    (contact_matcher_data['Industry'].isin(selected_industries))
]

# Segment data by industry and customer within the filtered data
industry_customer_metrics = filtered_data.groupby(['Order Date', 'Country', 'Industry', 'Customer']).agg({
    'Sales': 'sum',
    'Profit': 'sum',
    'Discount': 'median',
    'Quantity': 'sum',
    'Profit Percentage': 'median'
}).reset_index()

# Calculate Customer Lifetime Value (CLV) as total profit per customer
industry_customer_metrics['CLV'] = industry_customer_metrics['Profit']

# Calculate overall profit percentage for each customer within each industry
industry_customer_metrics['Overall Profit Percentage'] = (industry_customer_metrics['Profit'] / industry_customer_metrics['Sales']) * 100

# Display the first few rows of the segmented data for reference
industry_customer_metrics.head()
```
